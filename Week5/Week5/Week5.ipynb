# MatPlotLib in Python

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

## Simple Plots: plt.plot(X, Y)

```python
plt.plot([1, 2, 3, 4]) # considers the x-axis in range 0 - 3 by default 
plt.show()
```

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16])
plt.show()
```

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], linestyle="dotted", marker="^", color="green") #try [o, --, *, s, ^, +, _] etc.
plt.axis([0, 6, 0, 20])                     #first x-axis range followed by y-axis range
plt.show()
```

```python
plt.plot([0, 1, 2, 3, 4], [0, 1, 4, 9, 16], 'bo-') #try [o, --, *, s, ^, +, _] etc.
plt.axis([0, 6, 0, 20])                     #first x-axis range followed by y-axis range
plt.show()
```

## A little bit NumPy 

```python
# in order to get evenly distributed list between a given range, we'll use numpy's arange method or linspace method
x = np.arange(0, 1, 0.1)
y = np.linspace(0, 0.9, 10)
print(x)
print(y)
print(list(x)==list(y))
```

### Plotting Functions

```python
x = np.arange(0, 2*np.pi, 0.1)
y = np.sin(x)
y
```

```python
plt.plot(x,y)
plt.show()
```

```python
x = np.arange(0, 4, 0.1)
plt.plot(x, x**2)
plt.show()
```

### Plotting Multiple Functions in One Plot

```python
x = np.arange(0, 10, 0.1)
plt.plot(x, np.sin(x), x, np.cos(x))
plt.show()
```

```python
x = np.arange(0, 5, 0.2)
plt.plot(x, x, 'ro-', x, x**2, 'bs-', x, x**3, 'g^-')
plt.show()
```

### Plotting Multipe Functions in Multiple Plot

```python
fig, axs = plt.subplots(1, 1, figsize = (15, 5))
print(fig)
print(axs)
# print(type(axs))
# print(axs[0,0])
```

```python
x = np.arange(0, 5, 0.2)

plt.subplot(2, 2, 1)
plt.plot(x, x, 'r--')
plt.subplot(2, 2, 2)
plt.plot(x, x**2, 'bo')
plt.subplot(2, 2, 3)
plt.plot(x, x**3, 'ys')

plt.show()
```

```python
x = np.arange(0, 5, 0.2)

fig, axs = plt.subplots(1, 3, figsize = (15,5))
axs[0].plot(x, x, 'r--')
axs[1].plot(x, x**2, 'bo')
axs[2].plot(x, x**3, 'ys')

plt.show()
```

### Naming the Axes

```python
t = np.arange(1, 10, 0.2)
plt.figure(figsize = (10, 10))
plt.plot(t, np.sin(t), 'ro-')
plt.title("The Sine Graph", fontsize = 30, color = 'blue')
plt.xlabel("Time", fontsize = 20, color = 'purple')
plt.ylabel("Sin", fontsize = 20, color = 'red')
plt.show()
```

### Figure and Axes Sizes

```python
t = np.arange(1, 10, 0.1)

plt.figure(figsize=(10, 5), dpi = 100) #figure size in inches and resolution as dpi

plt.plot(t, np.sin(t), linestyle = 'dotted', linewidth = 5, marker = 'o') #'ko--'

plt.xlim(0.5,9.5) #x-axis limit
plt.xticks([2,4,6,8]) #ticks on the x-axis
plt.ylim(-2,2)
plt.yticks([-1,0,1])

plt.title("The Sine Graph", fontsize = 20, color = 'b')
plt.xlabel("Time (s)", fontsize = 20, color = 'purple')
plt.ylabel("Sin(t)", fontsize = 20, color = 'red')

plt.grid(True) #shows the grid

plt.show()
```

## Polar Plots

```python
theta = np.arange(0, 2 * np.pi + 0.001, 0.001)
r = np.sin(15*theta)
plt.figure(figsize = (10, 10))
plt.subplot(projection='polar')
plt.plot(theta, r, linewidth = 5)
plt.yticks([0,1,2])
plt.show()
```

### Legends

```python
t = np.arange(1, 10, 0.2)
plt.figure(figsize=(10, 5))

plt.plot(t, np.sin(t), label='Sine', linewidth = 3)
plt.plot(t, np.cos(t), label='Cosine', linewidth = 3)

plt.title("Sine vs Cosine Graph", fontsize = 20, color = 'b')
plt.xlabel("Time (s)", fontsize = 20, color = 'purple')
plt.legend(loc = 'lower left',  fontsize = 20) #try center, left, right, lower, upper
plt.savefig('sine-cosine.png')
plt.show()
```

## Histograms: plt.hist(X)

```python
np.random.rand(1)
```

```python
x = 50 + 20 * np.random.rand(1000) + 30 * np.random.rand(1000)
x
```

```python
x = 50 + 20 * np.random.rand(1000) + 30 * np.random.rand(1000)
plt.figure(figsize = (10, 5))
plt.hist(x, bins = 10, color = 'blue')
plt.xlim([50, 100])
plt.xticks([x for x in range(50, 100, 5)])
plt.xlabel('# of Students')
plt.ylabel('GPA')
plt.title('Histogram of GPAs')
plt.grid(True)
plt.show()
```

```python
#default histtype is bar
x = 50 + 20 * np.random.randn(1000)
y = 50 + 15 * np.random.randn(1000)
z = 50 + 10 * np.random.randn(1000)
fig, axs = plt.subplots(figsize = (10, 5))
axs.hist(x, bins = 20, color = 'blue', histtype = 'step', linewidth = 3)
axs.hist(y, bins = 20, color = 'red', histtype = 'step', linewidth = 3)
axs.hist(z, bins = 20, color = 'green', histtype = 'step', linewidth = 3)
plt.xlabel('# of Students')
plt.ylabel('GPA')
plt.title('Histogram of GPAs')
plt.grid(True)
plt.show()
```

```python
a = [70, 85, 95, 90, 56, 60, 65, 90, 75,74,74, 34, 23, 15, 5, 34, 34, 56]
plt.hist(a, bins = 5)
plt.show()
```

## Bars: plt.bar(X, Y)

```python
x = np.array([2000, 2005, 2010, 2015, 2020])
y = np.array([60, 65, 75, 70, 60])
plt.bar(x, y, width = 4, color = ['b', 'g', 'r', 'k'])
#plt.xticks(x)
plt.show()
```

```python
x = np.array([2000, 2005, 2010, 2015])
y = np.array([60, 65, 75, 70])
y_error = np.array([6,3,8,10])
plt.bar(x, y, width = 4, yerr = y_error)
plt.xticks(x)
plt.show()
```

## Pie Charts: plt.pie(X)

```python
shaarlar = 'Bishkek', 'Chui', 'Batken', 'Osh', 'Talas', 'Naryn', 'Yssyk-Kol', 'Djalal-Abad', 'Nookat'
data = [1000, 350, 100, 400, 200, 150, 300, 350, 550]
boluu = (0.1, 0, 0, 0, 0, 0.2, 0, 0, 0.3)

fig, axs = plt.subplots(figsize = (10, 10))
axs.pie(data, explode=boluu, labels=shaarlar, autopct='%1.1f%%', shadow=True, startangle=90, rotatelabels = True)

plt.show()
```

## Scatter Plot: plt.scatter(X, Y)

```python
GPA = 50 + np.random.rand(1000)
ORT = 110 + np.random.rand(1000)
fig, axs = plt.subplots(figsize=(10, 5))
axs.scatter(GPA, ORT)
plt.show()
```

```python
x = np.random.randint(1, 100, 100)
y = np.random.randint(1, 100, 100)
colors = ['r', 'b', 'g'] #np.random.rand(100)
area = np.pi * (15 * np.random.rand(100))**2
fig, axs = plt.subplots()
axs.scatter(x, y, s = area, c = colors ) #s is size array, c is a color array
plt.savefig('scatter.png') #saves the figure :)
plt.show()
```
